```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Dense, Dropout, LSTM, Conv1D, MaxPooling1D, Flatten, concatenate
)
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================
#  Paths
# ============================================================
base_path = r"C:\Users\NXTWAVE\Downloads\Disease Outbreak Prediction & Health Risk Monitoring System"
data_path = os.path.join(base_path, "archive", "Weather-related disease prediction.csv")

model_path = os.path.join(base_path, "hybrid_medican_model.keras")
scaler_path = os.path.join(base_path, "hybrid_medican_scaler.pkl")
yaml_path = os.path.join(base_path, "hybrid_medican_config.yaml")
json_path = os.path.join(base_path, "hybrid_medican_prediction.json")
csv_path = os.path.join(base_path, "hybrid_medican_result.csv")
visual_dir = os.path.join(base_path, "visuals")

os.makedirs(visual_dir, exist_ok=True)

# ============================================================
#  Load & Preprocess Data
# ============================================================
print("[INFO] Loading dataset...")
df = pd.read_csv(data_path)
print("[INFO] Shape:", df.shape)
print("[INFO] Columns:", df.columns.tolist())

# Keep numeric columns only
df = df.select_dtypes(include=[np.number]).dropna()
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)

X = scaled[:, :-1]
y = scaled[:, -1]

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Reshape for CNN-LSTM
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

joblib.dump(scaler, scaler_path)
print(f"[INFO] Scaler saved at: {scaler_path}")

# ============================================================
#  CNN-LSTM Model Builder
# ============================================================
def create_model(params):
    lr, dropout, lstm_units = params
    inputs = Input(shape=(X_train.shape[1], 1))

    cnn = Conv1D(filters=32, kernel_size=3, activation="relu")(inputs)
    cnn = MaxPooling1D(pool_size=2)(cnn)
    cnn = Flatten()(cnn)

    lstm = LSTM(int(lstm_units), return_sequences=False)(inputs)
    merged = concatenate([cnn, lstm])
    dense = Dense(64, activation="relu")(merged)
    dense = Dropout(dropout)(dense)
    output = Dense(1)(dense)

    model = Model(inputs, output)
    model.compile(optimizer=Adam(learning_rate=lr), loss="mse")
    return model

# ============================================================
#  Objective Function
# ============================================================
def objective_function(params):
    lr, dropout, lstm_units = params
    model = create_model([lr, dropout, lstm_units])
    model.fit(X_train, y_train, epochs=8, batch_size=32, verbose=0)
    preds = model.predict(X_test)
    return mean_squared_error(y_test, preds)

# ============================================================
#  Hybrid AIS + CSA Optimizer
# ============================================================
def hybrid_ais_csa(opt_iters=6, pop_size=5):
    print("[INFO] Running Hybrid AIS + CSA optimization...")

    lb = [1e-4, 0.1, 16]
    ub = [1e-2, 0.5, 128]

    # AIS: Initialize random antibodies (solutions)
    antibodies = np.random.uniform(lb, ub, (pop_size, len(lb)))
    fitness = np.array([objective_function(ab) for ab in antibodies])

    best_idx = np.argmin(fitness)
    best_antibody = antibodies[best_idx].copy()
    best_score = fitness[best_idx]

    memory = best_antibody.copy()  # CSA memory (best found so far)
    awareness = 0.1  # CSA awareness probability

    for t in range(opt_iters):
        print(f"[INFO] Iteration {t+1}/{opt_iters}...")
        
        # === AIS Phase: Clone & Mutate ===
        clones = []
        for ab in antibodies:
            n_clones = np.random.randint(1, 3)
            for _ in range(n_clones):
                mutation = ab + np.random.normal(0, 0.05, len(ab))
                mutation = np.clip(mutation, lb, ub)
                clones.append(mutation)
        clones = np.array(clones)

        all_candidates = np.vstack((antibodies, clones))
        all_fitness = np.array([objective_function(c) for c in all_candidates])

        best_idx = np.argmin(all_fitness)
        best_antibody = all_candidates[best_idx].copy()
        best_score = all_fitness[best_idx]

        # === CSA Phase: Follow best with random awareness ===
        new_antibodies = []
        for i in range(pop_size):
            if np.random.rand() > awareness:
                new_pos = antibodies[i] + np.random.rand() * (memory - antibodies[i])
            else:
                new_pos = np.random.uniform(lb, ub)
            new_pos = np.clip(new_pos, lb, ub)
            new_antibodies.append(new_pos)
        antibodies = np.array(new_antibodies)

        fitness = np.array([objective_function(ab) for ab in antibodies])
        cur_best_idx = np.argmin(fitness)
        if fitness[cur_best_idx] < best_score:
            best_score = fitness[cur_best_idx]
            best_antibody = antibodies[cur_best_idx].copy()
            memory = best_antibody.copy()

        print(f"[INFO] Best MSE so far: {best_score:.6f}")

    return best_antibody, best_score

# ============================================================
#  Run Hybrid Optimization
# ============================================================
best_params, best_score = hybrid_ais_csa(opt_iters=4, pop_size=5)
print(f"[INFO] Best Parameters (lr, dropout, lstm_units): {best_params}")
print(f"[INFO] Best MSE: {best_score:.6f}")

# ============================================================
#  Final Model Training
# ============================================================
model = create_model(best_params)
history = model.fit(X_train, y_train, epochs=40, batch_size=32, validation_split=0.2, verbose=1)
model.save(model_path)
print(f"[INFO] Model saved at: {model_path}")

# ============================================================
#  Evaluation
# ============================================================
y_pred = model.predict(X_test).flatten()
y_test = y_test.flatten()

mae = mean_absolute_error(y_test, y_pred)
try:
    rmse = mean_squared_error(y_test, y_pred, squared=False)
except TypeError:
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

metrics = {"MAE": float(mae), "RMSE": float(rmse), "R2": float(r2)}
print("[INFO] Metrics:", metrics)

# Save results
results_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
results_df.to_csv(csv_path, index=False)
print(f"[INFO] Results CSV saved at: {csv_path}")

with open(json_path, "w") as f:
    json.dump({"metrics": metrics, "best_params": [float(x) for x in best_params]}, f, indent=4)

with open(yaml_path, "w") as f:
    yaml.dump({"optimizer": "Hybrid AIS+CSA", "best_params": [float(x) for x in best_params]}, f)

# ============================================================
#  Visualizations (Save + Show)
# ============================================================

# --- Loss Graph ---
plt.figure()
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.title("Training and Validation Loss (Hybrid AIS+CSA)")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "hybrid_medican_loss_graph.png"))
plt.show()

# --- Comparison Graph ---
plt.figure()
plt.scatter(y_test, y_pred, alpha=0.7, color="green")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Prediction Comparison Graph (Hybrid AIS+CSA)")
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "hybrid_medican_comparison_graph.png"))
plt.show()

# --- Heatmap ---
corr_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
plt.figure(figsize=(5,4))
sns.heatmap(corr_df.corr(), annot=True, cmap="viridis")
plt.title("Correlation Heatmap (Hybrid AIS+CSA)")
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "hybrid_medican_heatmap.png"))
plt.show()

print("[] All graphs shown and saved successfully!")
print("[] Output Folder:", base_path)
```
